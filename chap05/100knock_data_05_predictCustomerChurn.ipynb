### knock 41 : Let's read the data and format the usage data.

```python
import pandas as pd
```

```python
customer = pd.read_csv('customer_join.csv')
uselog_months = pd.read_csv('use_log_months.csv')
```

Create data for the current month and one month prior data usage history only.


This is because many customers quit after only a few months, i.e., it is not possible to predict memberships within 5 months when using data from the previous 6 months.

```python
year_months = list(uselog_months["年月"].unique())
uselog = pd.DataFrame()
for i in range(1, len(year_months)):
    tmp = uselog_months.loc[uselog_months["年月"]==year_months[i]].copy()
    tmp.rename(columns={"count":"count_0"}, inplace=True)
    tmp_before = uselog_months.loc[uselog_months["年月"]==year_months[i-1]].copy()
    del tmp_before["年月"]
    tmp_before.rename(columns={"count":"count_1"}, inplace=True)
    tmp = pd.merge(tmp, tmp_before, on="customer_id", how="left")
    uselog = pd.concat([uselog, tmp], ignore_index=True)
uselog.head()
```

### knock 42 : Let's create the data of customers who left the membership in the month before the withdrawal.

Prepare data for the month prior to leaving the gym.

The gym can withdraw from membership at the end of the following month by submitting a withdrawal request by the end of the month. In other words, to prevent membership withdrawal, the month of withdrawal is August 2018, and the probability of submitting a withdrawal application in August is predicted from the data of July, one month prior to the month of withdrawal.

```python
from dateutil.relativedelta import relativedelta
```

```python
exit_customer = customer.loc[customer["is_deleted"]==1].copy()
exit_customer["exit_date"] = None
exit_customer["end_date"] = pd.to_datetime(exit_customer["end_date"])
for i in exit_customer.index:
    exit_customer.loc[i,"exit_date"] = exit_customer.loc[i,"end_date"] - relativedelta(months=1)
exit_customer["exit_date"] = pd.to_datetime(exit_customer["exit_date"])
exit_customer["年月"] = exit_customer["exit_date"].dt.strftime("%Y%m")
uselog["年月"] = uselog["年月"].astype(str)
exit_uselog = pd.merge(uselog, exit_customer, on=["customer_id", "年月"], how="left")
print(len(uselog))
exit_uselog.head()
```

```python
exit_uselog = exit_uselog.dropna(subset=["name"])
print(len(exit_uselog))
print(len(exit_uselog["customer_id"].unique()))
exit_uselog.head()
```

### knock 43 : Let's create data on continuing customers.

Continuing customers do not have a month of withdrawal, so any year and month data may be created.

```python
conti_customer = customer.loc[customer["is_deleted"]==0]
conti_uselog = pd.merge(uselog, conti_customer, on=["customer_id"], how="left")
print(len(conti_uselog))
conti_uselog = conti_uselog.dropna(subset=["name"])
print(len(conti_uselog))
```

Under-sampling because of the large and disproportionate amount of continuing customer data compared to the exiting data

```python
conti_uselog = conti_uselog.sample(frac=1, random_state=0).reset_index(drop=True)
conti_uselog = conti_uselog.drop_duplicates(subset="customer_id")
print(len(conti_uselog))
conti_uselog.head()
```

```python
predict_data = pd.concat([conti_uselog, exit_uselog],ignore_index=True)
print(len(predict_data))
predict_data.head()
```

### knock 44 : Let's create an enrollment period for the month you are projecting.

```python
predict_data["period"] = 0
predict_data["now_date"] = pd.to_datetime(predict_data["年月"], format="%Y%m")
predict_data["start_date"] = pd.to_datetime(predict_data["start_date"])
for i in range(len(predict_data)):
    delta = relativedelta(predict_data.loc[i, "now_date"], predict_data.loc[i, "start_date"])
    predict_data.loc[i, "period"] = int(delta.years*12 + delta.months)
predict_data.head()
```

### knock 45 : Let's remove missing values.

```python
predict_data.isna().sum()
```

Since end_date and exit_date only have values for exiting customers, continuing customers are missing values. So, only the missing data for count_1 is removed

```python
predict_data = predict_data.dropna(subset=["count_1"])
predict_data.isna().sum()
```

### knock 46 : Let's format variables of type string so that they can be processed.

```python
target_col = ["campaign_name", "class_name", "gender", "count_1", "routine_flg", "period", "is_deleted"]
predict_data = predict_data[target_col]
predict_data.head()
```

```python
predict_data = pd.get_dummies(predict_data)
predict_data.head()
```

Removing self-explanatory dummy variables.

```python
del predict_data["campaign_name_通常"]
del predict_data["class_name_ナイト"]
del predict_data["gender_M"]
predict_data.head()
```

### knock 47 : Let's create a predictive model for membership withdrawal using decision trees.

```python
from sklearn.tree import DecisionTreeClassifier
import sklearn.model_selection
```

```python
exit = predict_data.loc[predict_data["is_deleted"]==1]
conti = predict_data.loc[predict_data["is_deleted"]==0].sample(len(exit), random_state=0)

X = pd.concat([exit, conti], ignore_index=True)
y = X["is_deleted"]
del X["is_deleted"]
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,y, random_state=0)

model = DecisionTreeClassifier(random_state=0)
model.fit(X_train, y_train)
y_test_pred = model.predict(X_test)
print(y_test_pred)
```

Output result of 1 means withdrawal, 0 means predicted continuation

```python
results_test = pd.DataFrame({"y_test":y_test ,"y_pred":y_test_pred })
results_test.head()
```

A match between y_test and y_pred is a correct answer, and a mismatch is an incorrect answer.

### knock 48 : Let's evaluate the forecasting model and tune the model

```python
correct = len(results_test.loc[results_test["y_test"]==results_test["y_pred"]])
data_count = len(results_test)
score_test = correct / data_count
print(score_test)
```

```python
print(model.score(X_test, y_test))
print(model.score(X_train, y_train))
```

The output results show that the accuracy of the training data is higher than the prediction accuracy using the evaluation data, indicating a tendency toward overlearning.

Therefore, the depth of the tree structure of the decision tree is made shallower

```python
X = pd.concat([exit, conti], ignore_index=True)
y = X["is_deleted"]
del X["is_deleted"]
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,y, random_state=0)

model = DecisionTreeClassifier(random_state=0, max_depth=5)
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
print(model.score(X_train, y_train))
```

### knock 49 : Let's check the variables contributing to the model.

```python
importance = pd.DataFrame({"feature_names":X.columns, "coefficient":model.feature_importances_})
importance
```

The run results suggest that the number of times used in the previous month, length of enrollment, and whether or not it is a regular use are significant contributors, with a small contribution from the enrollment campaign.
